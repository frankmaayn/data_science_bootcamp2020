```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress as LR

#The file AnscombesQuartet.csv contains four data sets
df = pd.read_csv('AnscombesQuartet.csv')
df.head()
```

```python
#Get the x columns
xVal = [df['x1'],df['x2'],df['x3'],df['x4']]

#Get the y columns
yVal = [df['y1'],df['y2'],df['y3'],df['y4']]

#Sets the color
color = ['red','blue','yellow','green']

#Sets the title for each graph
gTitle = ['Graph1', 'Graph2', 'Graph3', 'Graph4']

#Sets the x,y label for each graph
xAxis = ['$X_1$','$X_2$','$X_3$','$X_4$']
yAxis = ['$Y_1$','$Y_2$','$Y_3$','$Y_4$']

#Sets the fontsize for label and title
f_SizeTitle = 12
f_SizeLabel = 10

#Create index for the values
index = 0


#abline formula
def abline(axs,slope,intercept):
    x_vals = np.array(axs.get_xlim())
    y_vals = intercept + slope * x_vals
    axs.plot(x_vals,y_vals, '--', label='lineReg')
    axs.legend()
```

```python
#Linear Regression - returns slope, y-intercept, r_value, p_value, stderr
slope, intercept, r_value,p_value, stderr =LR(xVal[0],yVal[0])  
```

```python
#Create a 2x2 scatter plot
fig, ax = plt.subplots(2,2)

for i in range(2):
    for j in range(2):
        if index < 4:
            slope, intercept, r_value, p_value, stderr = LR(xVal[index],yVal[index])
            ax[i,j].scatter(x=xVal[index], y=yVal[index], color=color[index], label='Values')
            ax[i,j].set_title(gTitle[index], fontsize = f_SizeTitle)
            ax[i,j].set_xlabel(xAxis[index], fontsize = f_SizeLabel)
            ax[i,j].set_ylabel(yAxis[index], fontsize = f_SizeLabel)
            ax[i,j].legend(loc='upper left')
            abline(ax[i,j],slope,intercept)
            ax[i,j].grid()
            index += 1
            
plt.gcf().set_size_inches(10,10)
plt.show()
```
