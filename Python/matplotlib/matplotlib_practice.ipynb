```python
import pandas as pd
import matplotlib.pyplot as plt
#Reads the CSV file
df = pd.read_csv('COVID19_line_list_data.csv', usecols=['age','death'])

#Drops missing values
df.dropna()

#Converts any non-zero into one -- indicates death
df['death'] = df['death'].apply(lambda x: 0 if x =='0' else 1)

#Converts column into floating point
df['death'].astype(float)
#Graph settings
plt.scatter(data=df, x='age', y='death')

#Graph label
plt.xlabel('Age')
plt.ylabel('Death')
plt.yticks([0,1], ['Surived','Death'])
plt.grid(True)

#Graph size
plt.gcf().set_size_inches(18,6)

#Graph title
plt.title('COVID19 Age vs. Death')

#show graph
plt.show()
```

```python
alive = df[df['death'] == 0].age
death = df[df['death'] == 1].age
#Plot the data
histog_out= plt.hist([alive,death], bins=10, label=['alive','death'])
#Add title
plt.title('COVID19 Alive vs. Death')
#Set the graph size and add legend
plt.gcf().set_size_inches(10,4)
plt.legend()
#Display
plt.show()
```

```python
import numpy as np
plotBins = 10

#Get the data from histogram
age = histog_out[1][1:]
num_alive = histog_out[0][0]
num_death = histog_out[0][1]
#Set bin
alive_bins = [1]* plotBins
death_bins = [3]* plotBins

#Plot
plt.scatter(x=age, y=alive_bins, s=num_alive*10,label='alive')
plt.scatter(x=age, y=death_bins, s=num_death*10, label='death')

#Set graph size
plt.gcf().set_size_inches(16,5)
#Set y-value ticks
plt.yticks([0,1,3,4],['','Survived','Death',''])
#Add legend
plt.legend()
#Display
plt.show()
```

```python
import numpy as np

def abline(axs,slope,intercept):
    x_vals = np.array(axs.get_xlim())
    y_vals = intercept + slope * x_vals
    axs.plot(x_vals,y_vals, '--', label='lineReg')
    axs.legend()
```
