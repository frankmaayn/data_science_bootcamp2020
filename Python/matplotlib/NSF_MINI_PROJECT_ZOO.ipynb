```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read the dataset
zoo_dataset = pd.read_csv('dataset_zoo.csv')

#Sort the dataset by class type
zoo_dataset = zoo_dataset.sort_values(['class_type'], ascending= True)
zoo_dataset
```

```python
class_type = np.array(zoo_dataset['class_type'])
hair = np.array(zoo_dataset['hair'])
feathers = np.array(zoo_dataset['feathers'])
eggs = np.array(zoo_dataset['eggs'])
milk = np.array(zoo_dataset['milk'])
airborne = np.array(zoo_dataset['airborne'])
aquatic = np.array(zoo_dataset['aquatic'])
predator = np.array(zoo_dataset['predator'])
toothed = np.array(zoo_dataset['toothed'])
backbone = np.array(zoo_dataset['backbone'])
breathes = np.array(zoo_dataset['breathes'])
venomous = np.array(zoo_dataset['venomous'])
fins = np.array(zoo_dataset['fins'])
legs = np.array(zoo_dataset['legs'])
tail = np.array(zoo_dataset['tail'])
domestic = np.array(zoo_dataset['domestic'])
catsize = np.array(zoo_dataset['catsize'])
```

```python
traits = [hair, feathers,eggs, milk, airborne, aquatic, predator, toothed, backbone, breathes, venomous, fins, legs, tail, domestic, catsize]
names = ['Hair', 'Feathers','Eggs', 'Milk', 'Airborne', 'Aquatic', 'Predator', 'Toothed', 'Backbone', 'Breathes', 'Venomous', 'Fins', 'Legs', 'Tail', 'Domestic', 'Catsize']
```

```python
fig, ax = plt.subplots(4,4)
index = 0
for i in range(4):
    for j in range(4):
        ax[i,j].scatter(class_type, traits[index], s=500)
        ax[i,j].set_ylabel(names[index], fontsize=30)
        ax[i,j].set_xlabel('Class Type', fontsize=30)
        ax[i,j].grid(c='r')
        index += 1
fig.set_size_inches(50,40)  
plt.show()
```
