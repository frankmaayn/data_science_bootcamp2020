```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

#Read the data set regression-data-set.csv

df = pd.read_csv('regression-data-set.csv')
df.head()
```

```python
#Convert the x and y columns into numpy arrays
xVal = np.array(df['x'])
yVal = np.array(df['y'])
```

```python
#Make a scatter plot of y as a function of x
plt.scatter(xVal,yVal)
```

```python
#Write a function that will perform polynomial fits of multiple degrees for asingle train/test split
#Function must return the MSE

def polynomial_fit(XVAL,YVAL,DEG):
    
    #train and tes split
    XTRAIN,XTEST,YTRAIN,YTEST = train_test_split(XVAL,YVAL)
    
    #perform a polyfit
    fit = np.polyfit(XTRAIN,YTRAIN,DEG)
    
    #perform a polyval
    ypredicted = np.polyval(fit,XTEST)
    
    #get the mean squared error and return
    MSE = mean_squared_error(YTEST,ypredicted)
    return MSE
```

```python
MSE = polynomial_fit(xVal,yVal, 1)
MSE
```

```python
#Perform 100 train/test splits and collect the errors as a function of degree
degree = 1
errors = []
for n in range(100):
    errors.append(polynomial_fit(xVal,yVal,degree))
```

```python
#plot histogram of the erros
plt.hist(errors)
```

```python
#Perform multiple train and test split with multiple degrees
degrees = np.linspace(1,15,15)
means = []
sigmas = []
for degree in degrees:
    errors = []
    for n in range(100):
       errors.append(polynomial_fit(xVal,yVal,degree)) 
    means.append(np.mean(errors))
    sigmas.append(np.std(errors))
```

```python
#Make a plot of the error as a function of degree
#Add the mean plus/minus 1 sigma
plt.plot(degrees, means, marker='o')
plt.plot(degrees, np.array(means)+1*np.array(sigmas))
plt.plot(degrees, np.array(means)-1*np.array(sigmas))

plt.yscale('log')
plt.gcf().set_size_inches(12,6)
plt.ylabel('Average MSE +/- SD', fontsize=16)
plt.xlabel("Degree of Fit", fontsize=16)
```

```python
#Fit the original data set to the favorite degree fit and plot it
plt.scatter(xVal,yVal, label='True Data')

fit = np.polyfit(xVal,yVal,9)
xplot = np.linspace(np.min(xVal),np.max(xVal),xVal.size)
yplot = np.polyval(fit,xplot)
plt.plot(xplot,yplot, label='Favorite fit', color='red')
plt.xlabel('X', fontsize=16)
plt.ylabel('Y', fontsize=16)
plt.legend(loc=2)
plt.gcf().set_size_inches(10, 5)
plt.show()
```
